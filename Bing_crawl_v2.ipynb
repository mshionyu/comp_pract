# これは何？

Bingの画像検索を用いて自動的にたくさんの画像をダウンロードする、Pythonプログラムです。


# 使い方

1. 下のコード中の、`our_dir = 'my_images'` の行の `my_images` を、ダウンロードする画像を保存したいフォルダの名前に変えます。その際、引用符を消さないこと。このフォルダは自動的に作られるので、事前に作っておく必要はありません。
2. 下のコード中の `keyword = '猫'` の行の `猫` を、画像の検索に使うキーワードに置き換えます。その際、引用符を消さないこと。複数のキーワードを半角スペースで区切って指定することもできます。
3. 下のコード中の、`class_label = '0'` の行の `0` を、その画像のラベルにする数字にします（初めて画像を集める場合は`0`から変更する必要なし）。その際、引用符を消さないこと。また、必ず半角の数字にして下さい。
4. 上記の書き換えが終わったら、セルを実行します。
    - うまくいけば、指定したフォルダに画像が落ちてくるはずです。
    - プログラムを実行し始めた後で、実行を止めたい場合は、上にある ■（中止）のボタンを押します。
5. 次の画像を集めたい場合は、検索キーワードとクラスラベルを変更して繰り返します。なお、クラスラベルは1ずつ増やすようにして下さい。

# その後やること

- （テキストを参考にして）AIに学習させるのに適さない画像を削除して下さい。


```python
from icrawler.builtin import BingImageCrawler
from pathlib import Path

root_path = Path('./')

###
out_dir = 'my_images' # 引用符の間に、ファイルを保存したいフォルダを記述する。
keyword = '猫'        # 猫 を消して、代わりに、自分が検索したい画像のキーワードを記述する。
class_label = '0'     # 引用符の間に、クラスのラベルを記述する。
###

path = root_path / out_dir / class_label
if not path.exists():
    path.mkdir(parents=True)

    crawler = BingImageCrawler(storage = {"root_dir" : path})
    crawler.crawl(keyword = keyword, max_num=1000)
else:
    print(f'{str(path)}は既に存在します。クラスラベルを確認してください。')
```
